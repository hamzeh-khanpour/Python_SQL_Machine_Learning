### Replace with mean
Missing values in attributes that have continuous data are best replaced using Mean value. We note that values in "Weight_kg" attribute are continuous in nature, and some values are missing. Therefore, write a code to replace the missing values of weight with the average value of the attribute.


````python
# Replace missing values in 'Weight_kg' with the column mean (Task 2)
if 'df' in globals():
    try:
        avg_weight = df['Weight_kg'].astype('float').mean(axis=0)
        df['Weight_kg'].replace(np.nan, avg_weight, inplace=True)
        print('Replaced missing Weight_kg with mean =', avg_weight)
    except Exception as e:
        raise RuntimeError('Error replacing Weight_kg missing values: ' + str(e))
else:
    raise RuntimeError('Dataframe `df` not found. Run the loading cells first.')
````

### Fixing the data types
Both "Weight_kg" and "Screen_Size_cm" are seen to have the data type "Object", while both of them should be having a data type of "float". Write a code to fix the data type of these two columns.


````python
# Fix data types for 'Weight_kg' and 'Screen_Size_cm' (Task 3)
if 'df' in globals():
    try:
        df[["Weight_kg","Screen_Size_cm"]] = df[["Weight_kg","Screen_Size_cm"]].astype("float")
        print('Converted Weight_kg and Screen_Size_cm to float (astype)')
    except Exception:
        # Try safer conversion using to_numeric with coercion
        df['Weight_kg'] = pd.to_numeric(df['Weight_kg'], errors='coerce')
        df['Screen_Size_cm'] = pd.to_numeric(df['Screen_Size_cm'], errors='coerce')
        print('Converted using to_numeric with coercion; non-convertible entries set to NaN')
else:
    raise RuntimeError('Dataframe `df` not found. Run the loading cells first.')
````

# Task - 4

### Data Standardization
The value of Screen_size usually has a standard unit of inches. Similarly, weight of the laptop is needed to be in pounds. Use the below mentioned units of conversion and write a code to modify the columns of the dataframe accordingly. Update their names as well.

```{math}
1 inch = 2.54 cm
1 kg   = 2.205 pounds
```


````python
# Data standardization: convert weight from kg to pounds and screen size from cm to inches (Task 4)
if 'df' in globals():
    try:
        # ensure numeric types before conversion
        df['Weight_kg'] = pd.to_numeric(df['Weight_kg'], errors='coerce')
        df['Screen_Size_cm'] = pd.to_numeric(df['Screen_Size_cm'], errors='coerce')
        # perform conversions and write to new columns
        df['Weight_pounds'] = df['Weight_kg'] * 2.205
        df['Screen_Size_inch'] = df['Screen_Size_cm'] / 2.54
        # drop original columns if they exist
        if 'Weight_kg' in df.columns:
            df.drop(columns=['Weight_kg'], inplace=True)
        if 'Screen_Size_cm' in df.columns:
            df.drop(columns=['Screen_Size_cm'], inplace=True)
        print('Converted Weight_kg->Weight_pounds and Screen_Size_cm->Screen_Size_inch')
    except KeyError as e:
        raise RuntimeError('Expected column missing: ' + str(e))
    except Exception as e:
        raise RuntimeError('Error during standardization: ' + str(e))
else:
    raise RuntimeError('Dataframe `df` not found. Run the loading cells first.')
````

### Data Normalization
Often it is required to normalize a continuous data attribute. Write a code to normalize the "CPU_frequency" attribute with respect to the maximum value available in the dataset.


````python
# Data normalization: normalize 'CPU_frequency' by its maximum value (Task - Normalization)
if 'df' in globals():
    try:
        # ensure numeric
        df['CPU_frequency'] = pd.to_numeric(df['CPU_frequency'], errors='coerce')
        max_val = df['CPU_frequency'].max()
        if pd.isna(max_val) or max_val == 0:
            raise RuntimeError('CPU_frequency max is NaN or zero; cannot normalize')
        df['CPU_frequency'] = df['CPU_frequency'] / max_val
        print('Normalized CPU_frequency by max value =', max_val)
    except KeyError:
        raise RuntimeError('Column CPU_frequency not found in dataframe')
    except Exception as e:
        raise RuntimeError('Error during normalization: ' + str(e))
else:
    raise RuntimeError('Dataframe `df` not found. Run the loading cells first.')
````

# Task - 5

### Binning
Binning is a process of creating a categorical attribute which splits the values of a continuous data into a specified number of groups. In this case, write a code to create 3 bins for the attribute "Price". These bins would be named "Low", "Medium" and "High". The new attribute will be named "Price-binned".


````python
# Create 3 bins for 'Price' named Low/Medium/High (Task 5)
if 'df' in globals():
    try:
        # ensure Price is numeric
        df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
        bins = np.linspace(df['Price'].min(), df['Price'].max(), 4)
        group_names = ['Low', 'Medium', 'High']
        df['Price-binned'] = pd.cut(df['Price'], bins, labels=group_names, include_lowest=True)
        print('Price-binned created. Counts:')
        print(df['Price-binned'].value_counts())
    except KeyError:
        raise RuntimeError('Column Price not found in dataframe')
    except Exception as e:
        raise RuntimeError('Error creating Price bins: ' + str(e))
else:
    raise RuntimeError('Dataframe `df` not found. Run the loading cells first.')
````

Also, plot the bar graph of these bins.


````python
# Plot a bar graph of the Price bins
if 'df' in globals():
    try:
        counts = df['Price-binned'].value_counts().reindex(['Low','Medium','High']).fillna(0)
        plt.bar(['Low','Medium','High'], counts)
        plt.xlabel('Price')
        plt.ylabel('count')
        plt.title('Price bins')
        plt.show()
    except Exception as e:
        raise RuntimeError('Error plotting Price bins: ' + str(e))
else:
    raise RuntimeError('Dataframe `df` not found. Run the loading cells first.')
````
